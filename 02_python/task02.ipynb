```python
price = int(input("税抜き価格を入力してください: ")) # 税抜き金額入力
priceIncludingTax = int(price * 1.1) # 消費税10% 含めた金額計算
postage = 350 # 送料

if price < 2000: # 2000円未満の場合は送料有り
    print(f"送料は{postage}円です")
    print(f"送料込税込価格は{priceIncludingTax + postage}円です")
else: # 2000円以上は送料無し
    print(f"送料は無料です")
    print(f"送料込税込価格は{priceIncludingTax}円です")
```
