```python
grades = [85, 90, 78, 92, 88]
num = len(grades)

total = sum(grades)
average = total/num
print(f"平均点: {average}")

max_num = max(grades)
print(f"最高点: {max_num}")

min_num = min(grades)
print(f"最低点: {min_num}")

target_student_list = list(filter(lambda x: x >= average, grades))
target_student_num = len(target_student_list)
print(f"平均以上の学生数: {target_student_num}")
```
