```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# インデックス0行0列の要素にアクセス（1行目1列目）
print(matrix[0][0])  # 出力: 1

# インデックス2行1列の要素にアクセス（3行目2列目）
print(matrix[2][1])  # 出力: 8
```

```python
modified_numbers = [i*10 if i % 2 == 0 else i * 2 for i in range(1, 6)]
print(modified_numbers)  # 出力: [2, 2, 6, 4, 10]
```

```python
# 各要素の2乗を格納するリスト
squares = [i**2 for i in range(1, 6)]
print(squares)  # 出力: [1, 4, 9, 16, 25]
```

```python
import copy

# 元のリスト（多次元リスト）
original_list = [[1, 2], [3, 4]]

# 浅いコピーを作成
shallow_copied_list = copy.copy(original_list)
# 深いコピーを作成
deep_copied_list = copy.deepcopy(original_list)

# 浅いコピーのリストを変更
shallow_copied_list[0][0] = 100
# 深いコピーのリストを変更
deep_copied_list[1][1] = 200

# 結果の出力
print("元のリスト:", original_list)             
print("浅いコピー:", shallow_copied_list)       
# → 浅いコピーの変更が元のリストに影響している

print("深いコピー:", deep_copied_list)         
# → 深いコピーの変更は元のリストには影響しない
```

```python
# 元のリスト
original = [1, 2, 3]

# スライスによるコピー
copy1 = original[:]
print(copy1)  # [1, 2, 3]

# list() 関数によるコピー
copy2 = list(original)
print(copy2)  # [1, 2, 3]

# copy() メソッドによるコピー
copy3 = original.copy()
print(copy3)  # [1, 2, 3]
```

```python
# 数値を格納したリスト
numbers = [1, 2, 3, 4, 5]
print(numbers) # [1, 2, 3, 4, 5]

# 文字列を格納したリスト
fruits = ["apple", "banana", "cherry"]
print(fruits) # ['apple', 'banana', 'cherry']

# 異なるデータ型を混在させたリスト
mixed = [1, "apple", 3.14, True]
print(mixed) # [1, 'apple', 3.14, True]

# 空のリスト
empty_list = []
print(empty_list) # []
```
