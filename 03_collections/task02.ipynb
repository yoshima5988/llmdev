```python
from collections import Counter

word_list = []

while True:
    word = str(input("英単語を入力してください: "))
    if word == "":
        break
    word_list.append(word)

# 入力された英単語をABC順に並べ替え
word_list.sort()
print(f"入力した英単語：{word_list}")

# 全ての英単語を連結
all_word = "".join(word_list)

# アルファベット毎に集計
alphabet_count_list = Counter(all_word)

# アルファベット毎の件数出力
for alphabet in sorted(alphabet_count_list):
    print(f"{alphabet}が{alphabet_count_list[alphabet]}個ありました")


```
