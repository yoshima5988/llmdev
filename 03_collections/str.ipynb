```python
grades = [85, 90, 78, 92, 88]
num = len(grades)

total = sum(grades)
average = total/num
print(f"平均点: {average}")
```

```python
s = "にんてんどううどんてんに"
if s == s[::-1]: # 元の値とひっくり返した値が同値である
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")

s = "Pythonは便利だが遅い"
print(s[::-1])
if s == s[::-1]:
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")

s = "no devil lived on"
if s == s[::-1]:
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")
```

```python
s = "abcdefghijklmnopqrstuvwxyz"

print(s[0:10:2]) # 2個おき
print(s[1:20:3]) # 3個おき
print(s[-1:-10:-2]) # 後から
```

```python
s = "abcdefghijklmnopqrstuvwxyz"

# 前から
print(s[0:7])
print(s[21:26])

# 後から
print(s[-5:-1])
```

```python
s = "abcdefghijklmnopqrstuvwxyz"

# 前から
print(s[0]) # a
print(s[1]) # b
print(s[25]) # z

# 後から
print(s[-1]) # z
print(s[-3]) # x
print(s[-26]) # a
```
