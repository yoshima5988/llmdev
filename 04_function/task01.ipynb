```python
# バリデーションチェク
def validation(name):
    if not check_lnegth(name):
        return False
    if not check_blank(name):
        return False
    if not check_isalnum(name):
        return False
    return True

# 文字数チェック
def check_lnegth(name):
    name_len = len(name)
    if (5 <= name_len <= 15):
        return True
    print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
    return False

# 空白チェック
def check_blank(name):
    if ' ' in name:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    return True

# 英数字チェック
def check_isalnum(name):
    if name.isalnum():
        return True    
    print("エラー: ユーザー名は英数字のみ使用できます")
    return False

name = input("ユーザ名を入力してください: ")
is_validate = validation(name)
if is_validate:
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
